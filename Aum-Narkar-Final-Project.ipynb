## Modeling Portion of My Project

```python
#Loading in the dataframe
df = pd.read_csv('/Users/aum/Documents/DB Final Datasets/DB_final_data.csv')
```

```python
#Adding region column for models to use
df['region'] = (df.index // 1000).map(region_map)
df.head()
```

```python
#Dropping any rows where the target column is null
df = df[df['outcome'].notnull()]
```

```python
#Importing necessary libraries/elements
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.linear_model import LogisticRegression

#Initializing my target column and other data
X = df.drop(['outcome', 'date', 'citation_issued', 'warning_issued', 'arrest_made', 'frisk_performed', 'raw_race',
 'raw_posted_speed',
 'raw_actual_speed',
 'raw_search_reason',
 'raw_contraband_or_evidence',
 'raw_custodial_arrest_made',
 'raw_search_vehicle_description',
 'raw_result_of_contact_description',
 'county_name',
 'officer_id_hash',
 'reason_for_frisk',
 'reason_for_search',
 'raw_Ethnicity',
 'raw_Race',
 'raw_action_description',
 'raw_ResultOfStop'
 ], axis=1)

y = df['outcome']

```

```python
#Finding all categorical columns to encode
cat_cols = X.select_dtypes('object').columns.tolist()
cat_cols
```

```python
#Building encoder for categorical data
ohe = OneHotEncoder(sparse_output=False, handle_unknown = 'ignore')
transformer = make_column_transformer((ohe, cat_cols), remainder = 'passthrough',
                                     verbose_feature_names_out=False)
```

```python
#Train-test split to avoid overfitting
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#Calculating baseline accuracy to compare model results to
print(f'Baseline Accuracy for the Dataset: {round((y_test.value_counts(normalize=True)[0]*100),3)}%')
```

```python
#Initializing decision treee + building out pipeline
tree = DecisionTreeClassifier()
pipe = Pipeline([('transformer', transformer), ('model', tree)])
```

```python
#Setting parameters for GridSearch
params = {'model__max_depth': range(0, 15, 3), 
          'model__criterion' : ['entropy', 'gini']}
```

```python
#Initializing GridSearchCV
grid = GridSearchCV(pipe, param_grid=params, scoring='accuracy')
```

```python
#Fitting the GridSearch onto train data
grid.fit(X_train, y_train)
```

```python
#Extracting the best parameters
grid.best_params_
```

```python
#Scoring the best model based on accuracy
grid.score(X_train,y_train)
```

```python
#Creating dataframe to calculate model accuracy
model_predictions = grid.predict(X_test)
acc_df = pd.DataFrame({"Model Predictions" : model_predictions, 'Actual Values': y_test})
acc_df['Correct?'] = (acc_df['Actual Values'] == acc_df['Model Predictions'])
print(f'Decision Tree Accuracy: {(round(acc_df['Correct?'].value_counts(normalize=True)[0], 5)*100)}%')
```

```python
#Printing out ConfusionMatrix
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(grid, X_train, y_train)
```

```python
#Printing out dataframe of feature importances
pd.DataFrame({'Feature': grid.best_estimator_.named_steps["transformer"].get_feature_names_out(), 'Importance':grid.best_estimator_.named_steps["model"].feature_importances_}).sort_values(by='Importance', ascending=False).head(25)
```

### Multinomial Logistic Regressor

```python
#Building a new dataframe with relevant variables for analysis
lgr_df = df[['time', 'subject_age','subject_race', 'subject_sex','contraband_found','search_conducted','region', 'outcome']]
#Filling N/A values for contraband (since blank = no contraband)
lgr_df['contraband_found'] = lgr_df['contraband_found'].fillna(False)
#Dropping any null values since model cannot handle null values
lgr_df.dropna(inplace=True)
```

```python
#Cleaning all of the columns in the dataset, to avoid multiple data types within each column
for column in lgr_df.columns:
    if lgr_df[column].dtype == 'float64' or lgr_df[column].dtype == bool:
        lgr_df[column] = lgr_df[column].astype(int)

lgr_df.info()
```

```python
#Shows some class imbalance, but due to skewed nature of dataset can't really fix that (otherwise findings wouldn't be generalizable if we don't randomly sample)
lgr_df['outcome'].value_counts()
```

```python
#Setting my target column and other data
X_lgr = lgr_df.drop('outcome', axis=1)
y_lgr = lgr_df['outcome']
#Parsing for the categorical columns
lgr_cat_cols = X_lgr.select_dtypes('object').columns.tolist()
lgr_cat_cols
```

```python
#Encoding categorical columns
transformer = make_column_transformer((ohe, lgr_cat_cols), remainder = 'passthrough',
                                     verbose_feature_names_out=False)
```

```python
#Train-test split to avoid overfitting
X_lgr_train, X_lgr_test, y_lgr_train, y_lgr_test = train_test_split(X_lgr, y_lgr, test_size=0.4)
```

```python
from sklearn.preprocessing import LabelEncoder

#Encode string labels to integers
le = LabelEncoder()
#Converts labels to [0, 1, 2] for train and test sets
y_lgr_train = le.fit_transform(y_lgr_train) 
y_lgr_test = le.transform(y_lgr_test)  
```

```python
#Initializing logistic regressor
lgr = LogisticRegression(multi_class='multinomial', max_iter=1500)
```

```python
#Building pipeline
pipe3 = Pipeline([('transformer', transformer), ('model', lgr)])
```

```python
#Setting parameters for GridSearch, using np.linspace to generate class weights
weights = np.linspace(0.0,1.0,4)
params = {'model__solver': ['sag', 'saga'], 
          'model__class_weight': [{0:x, 1:1.0-x} for x in weights]}
```

```python
#Initializing GridSearch object
grid = GridSearchCV(pipe3, param_grid=params, scoring='accuracy')
```

```python
#Fitting the grid onto train data
grid.fit(X_lgr_train, y_lgr_train)
```

```python
#Extracting the best parameters
grid.best_params_
```

```python
#Scoring model based on accuracy
grid.score(X_lgr_train,y_lgr_train)
```

```python
#Building DataFrame to calculuate model accuracy and outputting it
model_predictions = grid.predict(X_lgr_test)
acc_df = pd.DataFrame({"Model Predictions" : model_predictions, 'Actual Values': y_lgr_test})
acc_df['Correct?'] = (acc_df['Actual Values'] == acc_df['Model Predictions'])
print(f'Multinomial Regression Accuracy: {(round(acc_df['Correct?'].value_counts(normalize=True)[0], 5)*100)}%');
```

```python
#Finding the baseline accuracy for the test set by finding the most common category
print(f'Baseline Accuracy for Test Set: {round((pd.DataFrame(y_lgr_test).value_counts(normalize=True)[1])*100, 3)}%')
```

```python
#Outputting accuracy DataFrame
acc_df
```

```python
#Printing out ConfusionMatrix
ConfusionMatrixDisplay.from_estimator(grid, X_lgr_train, y_lgr_train)
```

```python
#Outputting a DataFrame of features and their respective importance
pd.DataFrame({'Feature': grid.best_estimator_.named_steps["transformer"].get_feature_names_out(), 'Importance':grid.best_estimator_.named_steps["model"].coef_[0]}).sort_values(by='Importance', ascending=False).head(25)
```
